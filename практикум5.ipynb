##2

```python
df2 = pd.read_csv('titanic_csv.csv', delimiter=';')
df2 = df2.rename (
    columns = {
        'PassengerId' : 'passengerid',
        'Pclass' : 'pclass',
        'Survived' : 'survived',
        'Name' : 'name',
        'Sex'  : 'sex',
        'Age' : 'age',
        'SibSp'      : 'sibsp',
        'Parch' : 'parch',
        'Ticket' : 'ticket',
        'Fare' : 'fare',
        'Cabin' : 'cabin',
        'Embarked' : 'embarked'
    }
)
df2
```

##3

```python
df3 = pd.read_csv('https://gist.githubusercontent.com/zaryanezrya/8b4ef51c707cb16d5e88a44dc00a1bb2/raw/41230f49c6268e072dbf102672f670be256922ab/gistfile1.txt')
df3
```

## 4

```python
df = pd.merge(df2, df3, how = 'outer')
df=df.drop_duplicates()
df
```

## 5

```python
df = df.set_index('passengerid')
```

```python
df.sort_values(by=['passengerid'])
```

```python
df.info()
df.describe()
```

##7

```python
tmp=df.iloc[0]
df.iloc[0]=df.loc[2]
df.loc[2]=tmp
df.iloc[0]
```

## 8

```python
def f(x):
  if x == "female":
    return "f"
  if x == 'male':
    return "m"
  else: return 'uknown'

df['sex'] = df['sex'].map(f)
df
```

##9

```python
df_check = df.groupby('ticket')['name'].count() # в датафрейм поместим все номера билетов и количество людей, которые числятся за этими билетами
tickets=df_check[df_check>=6] # возьмем только те билеты, где количество людей >= 6
tickets
```

## 10

```python
df[df['ticket'].isin(tickets.index)] #фильтруем датафрейм и оставляем только строки с номерами билетов, которые есть в tickets
```

```python
peoples=df[df['name'].isin(df['name'])]
peoples
```

```python
together = df[df['cabin'].isin(peoples['cabin'])]
together
```

## 11

```python
df['birthyear'] = 1912-df['age']
df
```

##12

```python
#сперва сделаем датафрейм, который будет содержать в себе кабины и количество людей, которые были в них
cabins = df.groupby('cabin')['name'].count() #сгруппируем по cabin и выведем количество людей, которые были в них
cabins
```

```python
#создаем столбец companion
df['companion'] = df['cabin'].map(cabins) #берем информацию о кабине из cabins, и в companion присваиваем количество людей в этой кабине
df.sort_index()
```
