```python
import numpy
import matplotlib.pyplot as plt
import pandas
import scipy
import sympy
```

```python
def f(x):         #1 задание
  return 2/(sympy.tan(x))

d = (6 - 4)/30
p = [4+d*i for i in range(30)][1:]
dots = []

for i in range(29):
  dots.append(f(p[i]))
  print(f(p[i]))
```

```python
plt.plot(p, dots) #2
plt.legend([r'$y = \frac{2}{tg(x)}$'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции ' r'$y = \frac{2}{tg(x)}$', fontsize=15)
plt.show()
```

```python
#3
plt.plot(p, dots, 'o', color=(0.3, 0.5, 0.11))
plt.xlabel('x')
plt.ylabel('y')
plt.axis((4, 6, -6, 2))
plt.grid(visible=True, which='major', axis='both', color = 'blue', linewidth=0.4, alpha = 0.7)
plt.title('График функции ' r'$y = \frac{2}{tg(x)}$', fontsize=15)
plt.show()
```

```python
#4.1
uniform_sample = numpy.round(numpy.random.uniform(low = 0, high = 100, size = 100))#4
normal_sample = numpy.random.normal(loc = 50, scale = 10, size = 100).astype(int)

plt.hist(uniform_sample, 10, facecolor='g', alpha=0.75)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Histogram of uniform sample')

plt.axis([0, 100, 0, 20])
plt.grid(True)
plt.show()
```

```python
#4.2
plt.hist(normal_sample, 15, facecolor='r', alpha=0.75)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Histogram of normal sample')

plt.axis([0, 100, 0, 25])
plt.grid(True)
plt.show()
```

```python
#5.1
vals = [1, 2, 3, 4, 5]
data = numpy.round(numpy.random.uniform(low = 1, high = 5, size = 50))#4
data2 = [numpy.sum(data==1), numpy.sum(data==2), numpy.sum(data==3), numpy.sum(data==4), numpy.sum(data==5)]
print('1 встречается: ', numpy.sum(data==1))
print('2 встречается: ', numpy.sum(data==2))
print('3 встречается: ', numpy.sum(data==3))
print('4 встречается: ', numpy.sum(data==4))
print('5 встречается: ', numpy.sum(data==5))

plt.bar(vals, data2, color = ['b', 'r', 'g', 'y', 'c'])
plt.title('Распределение цифр от 1 до 5', fontsize = 15)
```

```python
#5.2
fig, ax = plt.subplots()
ax.pie(data2, labels=vals, colors = ['b', 'r', 'g', 'y', 'c'])
plt.title('Распределение цифр от 1 до 5', fontsize = 15)
```
