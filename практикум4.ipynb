## Инициализация библиотек

```python
import numpy
import numpy as np
import matplotlib.pyplot as plt
import pandas
import scipy
import sympy
from mpl_toolkits.mplot3d import Axes3D
```

## 1 задание

```python
def f(x):         #1 задание
  return 2/(sympy.tan(x))

d = (6 - 4)/50
p = [4+d*i for i in range(50)][1:]
dots = []

for i in range(49):
  dots.append(f(p[i]))
  print(f(p[i]))
```

## 2 задание

```python
plt.plot(p, dots) #2
plt.legend([r'$y = \frac{2}{tg(x)}$'])
plt.xlabel('x', fontsize=10)
plt.ylabel('y', fontsize=10)
plt.grid(True)
plt.title('График функции ' r'$y = \frac{2}{tg(x)}$', fontsize=15)
plt.show()
```

## 3 задание

```python
#3
plt.plot(p, dots, 'o', color=(0.3, 0.5, 0.11))
plt.xlabel('x')
plt.ylabel('y')
plt.axis((4, 6, -6, 2))
plt.grid(visible=True, which='major', axis='both', color = 'blue', linewidth=0.4, alpha = 0.7)
plt.title('График функции ' r'$y = \frac{2}{tg(x)}$', fontsize=15)
plt.show()
```

# 5 задание

## Столбцевая диаграмма

```python
#5.1
vals = [1, 2, 3, 4]
data = numpy.round(numpy.random.uniform(low = 1, high = 4, size = 50))#4
data2 = [numpy.sum(data==1), numpy.sum(data==2), numpy.sum(data==3), numpy.sum(data==4)]
print('1 встречается: ', numpy.sum(data==1))
print('2 встречается: ', numpy.sum(data==2))
print('3 встречается: ', numpy.sum(data==3))
print('4 встречается: ', numpy.sum(data==4))

plt.bar(vals, data2, color = ['b', 'r', 'g', 'y'])
plt.title('Распределение цифр от 1 до 5', fontsize = 15)
```

## Круговая диаграмма

```python
#5.2
fig, ax = plt.subplots()
p = ["{0:.1f}%".format(v / (sum(vals)) * 100) for v in vals]
p[0] = "1 - " + p[0]
p[1] = "2 - " + p[1]
p[2] = "3 - " + p[2]
p[3] = "4 - " + p[3]
ax.pie(data2, labels=p, colors = ['b', 'r', 'g', 'y'])
plt.title('Распределение цифр от 1 до 5', fontsize = 15)

```

```python
#6
#def func(x, y):
  #  return (x - 1.5)**2 + (y - 2)**2
x = np.arange(-5, 5, -.25)
y = np.arange(-5, 5, -.25)
x, y = np.meshgrid(x, y)
z = (x - 1.5)**2 + (y - 2)**2
fig = plt.figure()
ax = fig.gca(projection = '3d')
surf = ax.plot_surface(x, y, z, color=[1, 0, 0, 1], linewidth = 0, antialiased = True)
fig.colorbar(surf, shrink = 0.5, aspect = 5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x)')
ax.set_title('Трехмерный график')

plt.tight_layout()
plt.show()
```
