```python
import numpy
import matplotlib.pyplot as plt
import pandas
import scipy
import sympy
```

```python
def f(x):         #1 задание
  return 2/(sympy.tan(x))

d = (6 - 4)/30
p = [4+d*i for i in range(30)][1:]
dots = []

for i in range(29):
  dots.append(f(p[i]))
  print(f(p[i]))
```

```python
plt.plot(p, dots) #2
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции ' r'$y = \frac{2}{tg(x)}$')
plt.show()
```

```python
#3
plt.plot(p, dots, 'o', color=(0.3, 0.5, 0.11))
plt.xlabel('x')
plt.ylabel('y')
plt.axis((4, 6, -6, 2))
plt.grid(visible=True, which='major', axis='both', color = 'blue', ms = '22')
plt.title('График функции ' r'$y = \frac{2}{tg(x)}$')
plt.show()
```
